### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

```python
goodbooks[goodbooks['authors'].str.contains(',')]
```

```python
dictionary_authors_reviews = {}
for index, row in goodbooks.iterrows():
    lista = row['authors'].split(',')
    
    for author in lista:
        author = author.strip()
        
        if author not in dictionary_authors_reviews:
            dictionary_authors_reviews[author] = row['work_text_reviews_count']/len(lista)
        else:
            dictionary_authors_reviews[author] += row['work_text_reviews_count']/len(lista)
```

### 13. Find the author with the highest average rating according to the goodbooks datasets.

```python
author_ratings = {}
for index, row in goodbooks.iterrows():
    lista = row['authors'].split(',')
    for author in lista:
        author = author.strip()
        if author not in author_ratings:
            author_ratings[author] = []
        author_ratings[author].append(row['average_rating'])

author_ratings
```

```python
author_avg_rating={}

for key, value in author_ratings.items():
    author_avg_rating[key]  = round(sum(value)/len(value),3)

author_avg_rating
```
